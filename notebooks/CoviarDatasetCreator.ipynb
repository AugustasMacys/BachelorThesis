```python
import json
import os
from pathlib import Path
import ntpath
```

```python
METADATA_FILENAME = "metadata.json"
```

```python
metadata_paths = []
for path in Path(r'D:\deepfakes\data\train').rglob('*.json'):
    metadata_paths.append(path)
```

```python
def get_fake_videos_with_corresponding_original_videos():
    fake_list = list()
    original_list = list()
    for metadata_path in metadata_paths:
        with open(metadata_path) as f:
            folder = ntpath.dirname(metadata_path)
            data = json.load(f)
            for id in data:
                if data[id]["label"] == "FAKE":
                    if data[id]["original"]:
                        original_id = data[id]["original"]
                        postfix = original_id[:-4] + "_mpeg4" + ".mp4" 
                        original_full_path = os.path.join(folder, postfix)
                        original_list.append(original_full_path)

                        postfix = id[:-4] + "_mpeg4" + ".mp4" 
                        fake_list.append(os.path.join(folder, postfix))

    return fake_list, original_list
```

```python
fake, real = get_fake_videos_with_corresponding_original_videos()
```
