```python
import os
from glob import glob
import ntpath
import pandas as pd
import numpy as np
```

```python
private_ground_truth_dataframe = r"D:/deepfakes/aws/labels_updated.csv"
private_predicted_dataframe = r"D:\deepfakes\inference\final_efficient_net4_private_test_set.csv"
```

```python
truth_df = pd.read_csv(private_ground_truth_dataframe)
predicted_df = pd.read_csv(private_predicted_dataframe)
```

```python
truth_df
```

```python
predicted_df
```

```python
filenames_truth = truth_df.filename
filenames_predicted = predicted_df.filename
missing = np.setdiff1d(filenames_truth, filenames_predicted)
missing
```

```python
missing_df = truth_df[truth_df['filename'].isin(missing)]
```

```python
missing_df
```

```python
new_predicted_df = pd.concat([predicted_df, missing_df], ignore_index=True)
```

```python
new_predicted_df = new_predicted_df.sort_values(by=["filename"], ignore_index=True)
```

```python
new_predicted_df
```

## Loss

```python
truth_labels = truth_df.label
predicted_labels = new_predicted_df.label
```

```python
from sklearn.metrics import log_loss
loss = log_loss(truth_labels, predicted_labels)
loss
```
