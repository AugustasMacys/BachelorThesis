```python
import pandas as pd
import ntpath
import os
```

```python
real_training_dataframe = r"D:/deepfakes/dataframes_pairs/real_training_dataframe.csv"
fake_training_dataframe = r"D:/deepfakes/dataframes_pairs/fake_training_dataframe.csv"
```

```python
real_dataframe = pd.read_csv(real_training_dataframe)
```

```python
fake_dataframe = pd.read_csv(fake_training_dataframe)
```

```python
real_dataframe
```

```python
names_real = [ntpath.basename(x)[:-4] for x in real_dataframe.video_name]
```

```python
names_fake = [ntpath.basename(x)[:-4] for x in fake_dataframe.video_name]
```

```python
identifiers = range(0, 300, 20)
```

```python
names_real
```

```python
path_fake_frames = r"D:\deepfakes\training_fake_faces"
path_real_frames = r"D:\deepfakes\training_real_faces"

resulted_list_real = []
for name in names_real:
    for prefix in identifiers:
        filename = name + '_' + str(prefix) + '.png'
        path_final = os.path.join(path_real_frames, name, filename)
        resulted_list_real.append(path_final)
        
resulted_list_fake = []
for name in names_fake:
    for prefix in identifiers:
        filename = name + '_' + str(prefix) + '.png'
        path_final = os.path.join(path_fake_frames, name, filename)
        resulted_list_fake.append(path_final)
```

```python
real_dataframe = pd.DataFrame(resulted_list_real)
fake_dataframe = pd.DataFrame(resulted_list_fake)
```

```python
fake_dataframe.rename(columns={0 :'image_path'}, inplace=True)
real_dataframe.rename(columns={0 :'image_path'}, inplace=True)
```

```python
fake_dataframe["label"] = 1
```

```python
real_dataframe["label"] = 0
```

```python
fake_dataframe.to_csv(r"D:/deepfakes/dataframes_pairs/fake_crops.csv", index=False)
real_dataframe.to_csv(r"D:/deepfakes/dataframes_pairs/real_crops.csv", index=False)
```
